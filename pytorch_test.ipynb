```python
def batching(seq, batch_size, pad_ix):
    batched_seq = []
    _seq = []
    for i, s in enumerate(seq):
        _seq.append(s)
        if len(_seq) == batch_size:
            batched_seq.append(_seq)
            _seq = []
    pad_length = batch_size - len(_seq)
    batched_seq.append(_seq + [pad_ix] * pad_length)
    return batched_seq          
```
